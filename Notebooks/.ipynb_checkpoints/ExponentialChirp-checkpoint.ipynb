# <br> **Exponential Chirp** 
## Generating stimulus .csv files for LED Zappelin' v2 
#### for both LED and Chrolis versions 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Stimuli Parameters

### General Parameters
##### Here users can change the duration and intensity of the three first part of the stimulus which precededs the chirp
##### Pay attention to the units

```python
t_On = 3000                 # On Duration in ms
On = 100                    # On Intensity value in %

t_Off = 3000                # Off Duration in ms
Off = 0                     # Off Intensity value in %

t_Mid = 3000                # Median Duration in ms
Mid = 50                    # Median Intensity value in %
```

### Chirp Parameters
#### Here users can modify the chirp features. 
#### Pay attention to the units

```python
StartFrequency = 0          # Chirp begining frequency in Hz
EndFrequency = 10           # Chirp final frequency in Hz
Duration = 15               # Chirp Duration in s
sampling = 250              # Number of values per second
Sampling = Duration * sampling 
```

# Generating the chirp stimulus

### Linear frequency increase modulation


```python
ExpTime = np.linspace(0, Duration*2*np.pi, Sampling) # 10 s
ExpModulation = np.exp(np.linspace(StartFrequency, np.log(EndFrequency), Sampling))
ExpSinusoid = np.sin(ExpTime*ExpModulation)*Mid+On/2
Exponential_Chirp = np.around(ExpSinusoid).astype(int)
```

```python
fig,ax1 = plt.subplots(1,1, figsize = (20,5))
ax2 = ax1.twinx()

ax1.plot(ExpTime,Exponential_Chirp,'r',linewidth = 1)
VLines = np.linspace(0,Duration*2*np.pi,Duration+1)
ax1.vlines(VLines, 0,100,'k','--',linewidth = 2.5)
ax2.plot(ExpTime,ExpModulation,'b',linewidth = 3)
```

### Setting the Stimulus file parameter
#### Here users should only concern themselves with the Trigger Mode part of the stimulus. 
<br></br>
If Trigger[0] is set to "0", the stimulation will only output one trigger signal at the beggining of the stimulation. Nothing else is then required from the Trigger array.
<br></br>
If Trigger[0] is set to "1" (by default), a trigger signal will be output at the beginning of the stimulation, then every t = Trigger[1] (in microseconds). The rest of the array can be ignored.
<br></br>
If Trigger[0] is set to a n higher number, this n number will corresponds to n custom triggers users can set to roll in sequence. n entries need then to be set up to Trigger[n+1]. The rest of the array can be ignored.

```python
resolution = 1000/sampling
Resolution = np.zeros(len(chirp_Linear))
Resolution[0] = resolution*1000

length = len(chirp_Linear)
nEntries = np.zeros(len(chirp_Linear))
nEntries[0] = length

Trigger = np.zeros(len(chirp_Linear))
Trigger[0] = 1                              # Trigger Mode
Trigger[1] = 1000000                        # Trigger Duration in microseconds
```

### Filling up the arrays
#### Here users decide what to fill the LED channel arrays with

```python
dict_Stim_Chirp = {'LED01': chirp_Linear,
                   'LED02': chirp_Linear,
                   'LED03': chirp_Linear,
                   'LED04': chirp_Linear,
                   'LED05': chirp_Linear,
                   'LED06': chirp_Linear,
                   'LED07': chirp_Linear,
                   'LED08': chirp_Linear,
                   'LED09': chirp_Linear,
                   'LED10': chirp_Linear,
                   'LED11': chirp_Linear,
                   'LED12': chirp_Linear,
                   'Resolution': Resolution,
                   'nEntries': nEntries,
                   'Trigger': Trigger 
                   }
df_Stim_Chirp = pd.DataFrame(dict_Stim_Chirp)
df_Stim_Chirp
```
