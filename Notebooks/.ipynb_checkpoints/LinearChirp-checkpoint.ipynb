## <br> **Linear Chirp Stimulation**
### Geneating stimulus for LED Zappelin v2 (LED or Chrolis versions)<br>


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Stimuli Parameters

### General Parameters

```python
t_On = 3000                 # On Duration in ms
On = 100                    # On Intensity value in %

t_Off = 3000                # Off Duration in ms
Off = 0                     # Off Intensity value in %

t_Mid = 3000                # Median Duration in ms
Mid = On/2                  # Median Intensity value in %
```

### Chirp Parameters

```python
StartFrequency = 0          # Chirp begining frequency in Hz
EndFrequency = 10           # Chirp final frequency in Hz
Duration = 15               # Chirp Duration in s
sampling = 250              # Number of values per second
Sampling = Duration * sampling 
```

# Chirp Stimuli

### Linear frequency increase modulation

```python
LinTime = np.linspace(0, Duration*2*np.pi, Sampling) # 10 s
LinModulation = np.linspace(StartFrequency, EndFrequency, Sampling)
LinSinusoid = np.sin(LinTime*LinModulation)*Mid+50
Linear_Chirp = np.around(LinSinusoid).astype(int)
```

```python
fig,ax1 = plt.subplots(1,1, figsize = (20,5))
ax2 = ax1.twinx()

ax1.plot(LinTime,Linear_Chirp,'r',linewidth = 1)
VLines = np.linspace(0,Duration*2*np.pi,Duration+1)
ax1.vlines(VLines, 0,100,'k','--',linewidth = 2.5)
ax2.plot(LinTime,LinModulation,'b',linewidth = 3)
```

# Stimulus creation

```python
chirp_On = np.zeros(int(t_On/1000*sampling))
chirp_On.fill(On)
chirp_Off = np.zeros(int(t_Off/1000*sampling))
chirp_Off.fill(Off)
chirp_Mid = np.zeros(int(t_Mid/1000*sampling))
chirp_Mid.fill(Mid)

chirp_Linear = np.hstack((chirp_On,chirp_Off,chirp_Mid,Linear_Chirp))
```

```python
resolution = 1000/sampling
Resolution = np.zeros(len(chirp_Linear))
Resolution[0] = resolution*1000

length = len(chirp_Linear)
nEntries = np.zeros(len(chirp_Linear))
nEntries[0] = length

Trigger = np.zeros(len(chirp_Linear))
Trigger[0] = 1
Trigger[1] = 1000000
```

```python
dict_Stim_Chirp = {'LED01': chirp_Linear,
                   'LED02': chirp_Linear,
                   'LED03': chirp_Linear,
                   'LED04': chirp_Linear,
                   'LED05': chirp_Linear,
                   'LED06': chirp_Linear,
                   'LED07': chirp_Linear,
                   'LED08': chirp_Linear,
                   'LED09': chirp_Linear,
                   'LED10': chirp_Linear,
                   'LED11': chirp_Linear,
                   'LED12': chirp_Linear,
                   'Resolution': Resolution,
                   'nEntries': nEntries,
                   'Trigger': Trigger 
                   }
df_Stim_Chirp = pd.DataFrame(dict_Stim_Chirp)
df_Stim_Chirp
```
